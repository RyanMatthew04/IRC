```python
from rapidfuzz import fuzz, process

# Master database company names (can be loaded from Excel/CSV too)
master_names = ["Amazon services ltd", "Amazon food limited",
    "Tata Motors Limited",
    "Reliance Industries Ltd",
    "Infosys Technologies",
    "HDFC Bank",
    "Housing Development Finance Corporation",
    "Bajaj Auto Ltd",
    "Larsen & Toubro",
    "ITC Limited",
    "Wipro Technologies",
    "Mahindra & Mahindra"
]

# Incoming new company name
new_name = "amazon ltd"

# Get top 5 fuzzy matches
top_matches = process.extract(new_name, master_names, scorer=fuzz.token_sort_ratio, limit=5)

# Display results
print("Top 5 matches:")
for match, score, _ in top_matches:
    print(f"{match} → Similarity: {score}%")
```

```python
from faker import Faker
import random
import pandas as pd

fake = Faker()

# Suffixes and abbreviations
suffixes = ['Ltd', 'Inc', 'Corp', 'LLC', 'Pvt Ltd', 'Technologies', 'Group', 'Enterprises', 'Solutions', 'Holdings']

def generate_base_company():
    base = fake.company().split()[0]
    suffix = random.choice(suffixes)
    return f"{base} {suffix}"

def swap_letters(word):
    if len(word) < 3:
        return word
    i = random.randint(0, len(word)-2)
    chars = list(word)
    chars[i], chars[i+1] = chars[i+1], chars[i]
    return ''.join(chars)

def random_variant(name):
    parts = name.split()
    base = parts[0]
    suffix = parts[1] if len(parts) > 1 else ""

    variant_type = random.choice(['typo', 'reorder', 'suffix_change', 'insert_noise', 'pluralize'])

    if variant_type == 'typo':
        base = swap_letters(base)
    elif variant_type == 'reorder':
        return f"{suffix} {base}"
    elif variant_type == 'suffix_change':
        suffix = random.choice([s for s in suffixes if s != suffix])
    elif variant_type == 'insert_noise':
        base = base + random.choice(['&Co', 'Intl', 'Global', 'Systems'])
    elif variant_type == 'pluralize':
        base = base + 's' if not base.endswith('s') else base

    return f"{base} {suffix}".strip()

# Generate 1000 base names and unique variants
base_names = [generate_base_company() for _ in range(1000)]
variant_names = []
for name in base_names:
    for _ in range(5):
        variant_names.append(random_variant(name))

# Combine and label
df = pd.DataFrame({
    'name': base_names + variant_names
    
})


df.to_csv('dummy.csv',index= False)
```

```python
# Incoming new company name
new_name = "Jefferson Limited"
master_names = df['Company Name'].to_list()

# Get top 5 fuzzy matches
top_matches = process.extract(new_name, master_names, scorer=fuzz.token_sort_ratio, limit=5)

# Display results
print("Top 5 matches:")
for match, score, _ in top_matches:
    print(f"{match} → Similarity: {score}%")
```
